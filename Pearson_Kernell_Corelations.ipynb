```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/content/filled_data.csv', delimiter=',', decimal=',', parse_dates=['Date'])

# Round the values to the desired decimal places
data['Mean Temperature'] = data['Mean Temperature'].apply(lambda x: round(float(x), 2))
data['PM10'] = data['PM10'].apply(lambda x: round(float(x), 2))
data['SO2'] = data['SO2'].apply(lambda x: round(float(x), 2))

# Extract the 'Mean Temperature', 'PM10', and 'SO2' columns
mean_temp_values = data['Mean Temperature']
pm10_values = data['PM10']
so2_values = data['SO2']

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot PM10 vs Mean Temperature
ax1.scatter(mean_temp_values, pm10_values, marker='o', s=10, color='orange')
ax1.set_xlabel('Mean Temperature')
ax1.set_ylabel('PM10')
ax1.set_title('PM10 vs Mean Temperature')

# Plot SO2 vs Mean Temperature
ax2.scatter(mean_temp_values, so2_values, marker='o', s=10, color='blue')
ax2.set_xlabel('Mean Temperature')
ax2.set_ylabel('SO2')
ax2.set_title('SO2 vs Mean Temperature')

# Adjust spacing between subplots
plt.tight_layout()

# Save the figure as a single image file
plt.savefig('graph.png')

# Display the plots
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# Read the CSV file
data = pd.read_csv('/content/filled_data.csv', delimiter=',', decimal=',', parse_dates=['Date'])

# Round the values to the desired decimal places
data['Mean Temperature'] = data['Mean Temperature'].apply(lambda x: round(float(x), 2))
data['PM10'] = data['PM10'].apply(lambda x: round(float(x), 2))
data['SO2'] = data['SO2'].apply(lambda x: round(float(x), 2))

# Calculate Pearson correlation
pearson_corr = data[['Mean Temperature', 'SO2', 'PM10']].corr(method='pearson')

# Calculate Kendall correlation
kendall_corr = data[['Mean Temperature', 'SO2', 'PM10']].corr(method='kendall')

# Print Pearson correlation
print("Pearson Correlation:")
print(pearson_corr)

# Print Kendall correlation
print("\nKendall Correlation:")
print(kendall_corr)
```
